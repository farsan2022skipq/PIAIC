# Generators and creating own generators.

```python
# Generator expressions.
gen = (i for i in range(100) if i%2 == 0)
print(gen)
```

```python
# How to make Generator funcitons.
# Generators dont return things but yeild things.

def simple_gen():
    yield 'Hi'
    yield 'Hello'
    yield 'there  !'

for i in simple_gen():
    print(i)

```

```python
# finding the correct combination.
CORRECT_COMBO = (3,6,1)
found_combo = False

for c1 in range(10):
    if found_combo:
        break
    for c2 in range(10):
        if found_combo:
            break
        for c3 in range(10):
            if (c1,c2,c3) == CORRECT_COMBO:
                found_combo=True
                print(f"Found the correct combo {c1,c2,c3}")
                break
```

```python
# Now we will use the generator for the same code above.
CORRECT_COMBO = (3,6,1)

def combo_gen():
    for c1 in range(10):
        for c2 in range(10):
            for c3 in range(10):
                yield (c1,c2,c3)

for (c1,c2,c3) in combo_gen():
    print(c1,c2,c3)
    if (c1,c2,c3) == CORRECT_COMBO:
        print(f"Found the correct combo {c1,c2,c3}")
        break

```

# Multiprocessing Module.

```python
import multiprocessing

def span(num):
    print(f"Spanned {num}")
    
if __name__ == "__main__":
    for i in range(100):
        p = multiprocessing.Process(target=span, args=(i,))
        p.start()
        p.join()
```
